## 8

```python
import pandas as pd
eight = pd.read_excel('8.-Invoices-with-Merged-Categories-and-Merged-Amounts.xlsx')
(eight.set_index(['Order ID']) #Using the order ID as index because thats the one that won't explode
       .apply(lambda col: col.str.split('|')) #splitting by the delimiter '|' on the
       .explode(['Category', 'Amount']) #'Category' 'Amount' columns
       .reset_index()                # reset index from order ID to integars
       .reindex(eight.columns, axis=1))
```

## 7

```python
seven = pd.read_excel('7.-Medicine-Data-with-lumped-Quantity-and-Measure.xlsx')
seven
seven['Measure'] = seven['Quantity'].str.extract('([A-Za-z]+)', expand = True) #another column with just the alphabets
seven['Quantity']=seven['Quantity'].str.extract('(\d+)').astype(float) #remove the alphabets only
seven.loc['Total'] = pd.Series(seven['Quantity'].sum(), index=['Quantity']) #gives a total of the quantity column
seven
```

## 6

```python
six = pd.read_excel('6.-Hospital-Data-with-Mixed-Numbers-and-Characters.xlsx')
#six = six[six['Ending Balance (Physical Count)'].str.contains('\d') & six['Ending Balance (Physical Count)'].str.contains('\D')]
#above code was used to display rows with strings 
#change i to 1, S to 5,O to 0,o to 0
six[['Beginning Balance ','Quantity Received ','Quantity Dispensed ','Losses and Adjustments','Ending Balance (Physical Count)']]\
=six[['Beginning Balance ','Quantity Received ','Quantity Dispensed ','Losses and Adjustments','Ending Balance (Physical Count)']]\
.replace(['o','S', 'O','i','I','s'], ['0', '5', '0', '1', '1', '5'], regex=True)
#replace the wrong strings with the right ints in the selected columns
convert_dict = {'Beginning Balance ':float
                , 'Quantity Received ':float
                ,'Quantity Dispensed ':float
                , 'Losses and Adjustments':float
                ,'Ending Balance (Physical Count)':float}
six = six.astype(convert_dict) 
six.loc['Total'] = pd.Series(six[['Beginning Balance ', 'Quantity Received ',
       'Quantity Dispensed ', 'Losses and Adjustments',
       'Ending Balance (Physical Count)']].sum(), index=['Beginning Balance ', 'Quantity Received ',
       'Quantity Dispensed ', 'Losses and Adjustments',
       'Ending Balance (Physical Count)'])
 
#convert the float to int
six
```


## 5

```python
five = pd.read_excel('5.-Jumbled-up-Customers-Details.xlsx',header = None)
five.columns = ['test'] #change the column name
five.test = five.test.apply(lambda x: x.replace('Name', ''))
five.test = five.test.apply(lambda x: x.replace('Address', '|'))
five.test = five.test.apply(lambda x: x.replace('Age', '|'))
five.test = five.test.apply(lambda x: x.replace('Gender', '|'))
five = five.test.str.split("|",n=4, expand = True)
five.columns = ['Name','Address','Age','Gender']
five
```

## 4

```python
import pandas as pd
four = pd.read_excel('4.-Badly-Structured-Sales-Data-4.xlsx', header = [0,1], index_col=[0,1])
four.iloc[0] = four.iloc[0].fillna(method='ffill')
four = (four.stack([0,1])
        .rename_axis(['Order ID','Order Date','Ship Mode','Segment'])
        .reset_index(name='Sales'))

four
```
