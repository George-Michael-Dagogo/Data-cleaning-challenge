```python
seven = pd.read_excel('7.-Medicine-Data-with-lumped-Quantity-and-Measure.xlsx')
seven
seven['Measure'] = seven['Quantity'].str.extract('([A-Za-z]+)', expand = True) #another column with just the alphabets
seven['Quantity'] = seven['Quantity'].str.extract(r'(\d+)').astype(float)  # Remove alphabets, extract digits


seven.loc['Total'] = pd.Series(seven['Quantity'].sum(), index=['Quantity']) #gives a total of the quantity column
seven
```

```python
import pandas as pd

# Load the Excel file
six = pd.read_excel('6.-Hospital-Data-with-Mixed-Numbers-and-Characters.xlsx')

# Define columns to clean and replace invalid characters
columns_to_clean = [
    'Beginning Balance ',
    'Quantity Received ',
    'Quantity Dispensed ',
    'Losses and Adjustments',
    'Ending Balance (Physical Count)'
]

# Replace invalid characters in the specified columns
replacement_map = {
    'o': '0', 'O': '0',  # Replace 'o' and 'O' with '0'
    'S': '5', 's': '5',  # Replace 'S' and 's' with '5'
    'i': '1', 'I': '1'   # Replace 'i' and 'I' with '1'
}
six[columns_to_clean] = six[columns_to_clean].replace(replacement_map, regex=True)

# Convert the cleaned columns to float
six[columns_to_clean] = six[columns_to_clean].astype(float)

# Add a 'Total' row by summing up all numeric columns
six.loc['Total'] = six[columns_to_clean].sum()

# Convert the float columns to integers (optional if totals are not floats)
six[columns_to_clean] = six[columns_to_clean].astype(int)

six
```

```python
from io import StringIO
import pandas as pd
import numpy as np
from IPython.display import HTML

with open('Delmich dataset.csv', 'rb') as f:
    data = f.read() 

data1 = data.replace(b'|', b',').replace(b'?',b'').replace(b',,,,,,,',b'')
df = pd.read_csv(StringIO(data1.decode('latin-1')))

df.columns = df.columns.str.replace(' ', '')

df = df.map(lambda x: " ".join(x.split()) if isinstance(x, str) else x)
mask = df['First_Name'].astype(str).str.contains('\d', na=False)

df = df[(df != df.columns).all(axis=1)]

cols_to_shift = df.columns[df.columns.get_loc('First_Name')+1:]
df.loc[mask, 'First_Name'] = df.loc[mask, cols_to_shift[0]]
df.loc[mask, cols_to_shift] = df.loc[mask, cols_to_shift].shift(-1, axis=1)

# df['First_Name'] = df['First_Name'].str.strip()

mask1 = df['First_Name'] == ''

cols_to_shift2 = df.columns[df.columns.get_loc('First_Name')+1:]
df.loc[mask1, 'First_Name'] = df.loc[mask1, cols_to_shift2[0]]  
df.loc[mask1, cols_to_shift2] = df.loc[mask1, cols_to_shift2].shift(-1, axis=1)



condition = df['First_Name'].str.split().str.len() == 2

split_names = df.loc[condition, 'First_Name'].str.split()
# df = df[(df != df.columns).all(axis=1)]
df.loc[condition, 'Last_Name'] = split_names.str[1]

df.loc[condition, 'First_Name'] = split_names.str[0]
df['Student_ID'] = df['Student_ID'].str.replace(r'\D', '', regex=True).str.replace('2139', '139')

condition2 = df['Gender'].str.split().str.len() == 2
split_gender_age = df.loc[condition2, 'Gender'].str.split()
# df = df[(df != df.columns).all(axis=1)]
df.loc[condition2, 'Gender'] = split_gender_age.str[0]

df.loc[condition2, 'Age'] = split_gender_age.str[1]

condition3 = df['Gender'].astype(str).str.contains('\d', na=False)
temp = df.loc[condition3, 'Age']


df.loc[condition3, 'Age'] = df.loc[condition3, 'Gender']

df.loc[condition3, 'Gender'] = temp

df['Gender'] = df['Gender'].str.upper()

df['Age'] = df['Age'].str.replace('*', '')
df['Enrollment_Date'] = df['Enrollment_Date'].replace(['NA', ''], pd.NA)

df['Enrollment_Date'] = pd.to_datetime(df['Enrollment_Date'],format='mixed', dayfirst=True)

df = df.drop('Unnamed:8', axis=1)

display(HTML(df.to_html()))
```
